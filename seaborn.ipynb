```python
#numerical data plotting in seaborn
#relplot has options of plotting both categorical and numerical data
#relplot>> show the relationship between two variables
#replot >> relationship plot
df
```

```python
#show tip is varying wrt total_bill >> scatterplot/by default relplot makes scatter plot
sns.relplot(x = 'total_bill', y = 'tip',data = df) #relationship between two variables, by default it gives scatter plot,kind = 'line'
#insight >> as total bill increases tip increases
```

```python
sns.relplot(x = 'total_bill', y = 'tip',data = df, hue = 'smoker')#hue gives the different color set
#insight >>customer who smoked gave the highest tip
```

```python
sns.relplot(x = 'total_bill', y = 'tip',data = df, hue = 'size' )
```

```python
sns.relplot(x = 'total_bill', y = 'tip',data = df, style = 'size' )#style gives the different representation of the column value
```

```python
sns.relplot(x = 'total_bill', y = 'tip',data = df, style = 'smoker', hue = 'time' )
```

```python
#for categorical s >> catplot
#plot between data and total bill, day is a categorical value
sns.catplot(x = 'day', y = 'total_bill', data = tips)
```

```python
#insights >> many people arrives on weekends on sat sun
#in cat plot,few datapoints are overlaping to see it separately  use swarm plot

sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'swarm')
```

```python
#boxplot>> to see the outliers
sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'box')
```

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'violin')
```

```python
#boxen plot is made from box plot
sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'boxen')
```

```python
sns.barplot(x = 'day', y = 'total_bill', data = tips)
#how total bill(aggregated) varies with different data
```

```python
sns.jointplot(x = tips.total_bill, y = tips.tip) #draw a plot of two variables with distribution, bivarent plot
```

```python
# joints plot >> to understand relationship and also distribution of two variables

sns.jointplot(x = tips.total_bill,y = tips.tip,kind = 'reg' ) #reg >> regression line to represent relationship
```

```python
sns.jointplot(x = tips.total_bill, y = tips.tip, kind = 'hex')
```

```python
sns.jointplot(x = tips.total_bill, y = tips.tip, kind = 'kde')#kernel density estimator>> contour>> metrological department
```
